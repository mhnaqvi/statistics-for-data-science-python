## Task 5: Use the appropriate tests to answer the questions provided.

H0 - No significant difference between median values of houses bounded and not bounded by charles river
H1 - Significant difference between median values of houses bounded and not bounded by charles river

```python
df.loc[(df['CHAS'] == 0), 'CHAS_T'] = 'FAR'
df.loc[(df['CHAS'] == 1), 'CHAS_T'] = 'NEAR'
df.head(5)
```

Is there a difference in Median values of houses (MEDV) for each proportion of owner occupied units built prior to 1940 (AGE)? (ANOVA)

```python
df.head()
```

```python
farMEDV = df[df['CHAS_T'] == 'FAR']['MEDV']
nearMEDV = df[df['CHAS_T'] == 'NEAR']['MEDV']
scipy.stats.ttest_ind(farMEDV, nearMEDV, equal_var = True)
```

INTERPRETITION OF THE RESULT -
The P-value is less than 0.05 - H0 is rejected and H1 is accepted


H0 - No difference in median values units built prior to 1940
H0 - Yes difference in median values units built prior to 1940



```python
from statsmodels.formula.api import ols
lm = ols('MEDV ~ AGE', data = df).fit()
table = sm.stats.anova_lm(lm)
print(table)
```

INTERPRETITION OF THE RESULT -

The P-value is less than 0.05 H0 is accepted
